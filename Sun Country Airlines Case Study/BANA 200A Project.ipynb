```python
from sklearn.cluster import KMeans as KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
k = 5
data = pd.read_csv(r"C:\Users\Ethan\Downloads\Clustering Data.csv")
data.head(10).style.background_gradient(cmap='plasma').set_properties(**{'font-family': 'Segoe UI'})
```

(2) KMeans won't work with the data as is, so we'll have to clean it up a bit. Let's remove the unique user IDs to protect privacy of our data sources: This means we want to remove the uid & PNRLocatorID. Let's save it under a new name, data_2

UFLY rewards, Booking channel, Age, days_pre_booked

```python
data_2 = pd.DataFrame(data)
data_2 = data_2.drop('uid', axis = 1)
data_2 = data_2.drop('PNRLocatorID', axis = 1)


```

```python
len(data_2.columns)
```

```python
data_2_np = data_2.values
data_2_np
```

```python
kmeans = KMeans(n_clusters=k)
kmeans.fit(data_2_np)
cluster_assignments = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
cluster_assignments
cluster_centers
```

```python
labels = kmeans.labels_
plt.scatter(data_2_np[:, 0], data_2_np[:, 1], c=labels)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```
